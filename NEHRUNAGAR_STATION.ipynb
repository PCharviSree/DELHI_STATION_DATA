```python
#droping 'To Date' column from the dataframe
df.drop('To Date',axis=1,inplace=True)
```

```python
#creating list of columns from the dataset
cols = list(df.columns)
```

```python
#replacing None values to Nan values
df.replace('None',np.nan,inplace=True)
```

```python
#chekcing data by picking randomly with help of sample method
df[cols[1:-1]].sample(100)
```

```python
#replacing ' ' with nan values
df.replace(' ',np.nan,inplace=True)
```

```python
#checking data types of each column
df.dtypes
```

```python
#changing data type of colums from index 1 to -1 to float
for i in cols[1:-1]:
    df[i]=df[i].astype('float64')
```

```python
df
```

```python
years = []
for i in df['From Date']:
    years.append(i.year)
df['year'] = years
df
```

```python
months = []
for i in df['From Date']:
    months.append(i.month)
df['month'] = months
df
```

```python
days = []
for i in df['From Date']:
    days.append(i.day)
df['day'] = days
df
```

```python
#droping rows were month as empty string
df.drop(df[df['month']==''].index,axis=0,inplace=True)
```

```python
#checking datastype of month column to int
df['month']=df['month'].astype('int64')
```

```python
#creating quarters list and adding quarter column to dataset
quarters=[]
for i in df['month']:
    if i>=1 and i<=3:
        quarters.append('Q1')
    elif i>=4 and i<=6:
        quarters.append('Q2')
    elif i>=7 and i>=9:
        quarters.append('Q3')
    else:
        quarters.append('Q4')
df['Quarter']=quarters
#checking dataframe after adding quarter column
df.head()
```
